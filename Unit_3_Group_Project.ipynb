```python
covid_df_wyoming.plot(x='date',y='cases', color='purple')
```

```python
covid_df_vermont.plot(x='date',y='cases')
```

```python
x=covid_df_wyoming['date']
x_2=covid_df_vermont['date']
y= covid_df_wyoming['cases']
y_2=covid_df_vermont['cases']
fig,ax=plt.subplots()

ax.plot(x,y, color='purple')
ax.plot(x_2,y_2)
plt.show()
```

**FINDINGS** 

Wyoming = Purple
Vermont = Blue 

At first, in Wyoming and Vermont the cases rose steadily. However, as time passed Wyoming exceeded Vermont in cases but ultimately Vermont exceeded with cases towards the end. This lets us conclude that with similar population status, there can be similar trends in the contraction of the virus.
